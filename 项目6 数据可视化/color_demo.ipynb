```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import font_manager

font_path = './font/SimHei.ttf'  # 此处改为自己字体的路径
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams['font.family'] = prop.get_name()
plt.rcParams['mathtext.fontset'] = 'cm'  # 'cm' (Computer Modern)
plt.rcParams['axes.unicode_minus'] = False

# 准备数据
x = np.arange(2011,2020)
y = [10.15,12.11,10.89,12.41,10.59,13.07,11.77,9.84,9.14]
c1=['C9','C8','C7','C6','C5','C4','C3','C2','C1']
c2=['C7','C7','C7','C7','C7','C3','C7','C7','C7']

# 正确显示中文
mpl.rcParams['font.family'] = 'SimHei'

# 初始化画布
fig = plt.figure(figsize=(18,5),dpi=100)
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# 绘制数据
ax1.bar(x,y,color=c1)
ax2.bar(x,y,color='C7')
ax3.bar(x,y,color=c2)

ax2.set_yticks([])
ax3.set_yticks([])

# 设置标题
ax1.set_title('(a)')
ax2.set_title('(b)')
ax3.set_title('(c)')
fig.suptitle('2011-2019年 上海市出生人口（单位：万）');
```
