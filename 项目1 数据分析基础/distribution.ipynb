```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
n = int(input("How many times have you flipped a coin?"))  # 做某件事情的次数
p = 0.5 # 做某件事情成功的概率
X = np.arange(0, n+1,1)   # arange用于生成一个等差数组，arange([start, ]stop, [step, ])
X   # X 是成功的次数
```

```python
# 求对应分布的概率: 概率质量函数 (PMF)
# pmf 函数返回一个列表，列表中每个元素表示随机变量中对应值的概率

pList = stats.binom.pmf(X, n, p)   # (X:事件成功的次数，n:事件总次数，p:每次成功的概率)
pList
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
fig,ax = plt.subplots(dpi=100)
plt.plot(X, pList, marker='o',linestyle='None')  # linestyle：线条的形状，值 None 表示不显示连接各个点的折线
plt.vlines(X, 0, pList,linestyles='dashed')  # vline(x坐标值, y坐标最小值, y坐标值最大值)
plt.xlabel('随机变量：抛硬币正面朝上次数')
plt.ylabel('概率')
ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=1))
plt.title('二项分布：n=%i,p=%.2f' % (n,p))
plt.show()
```

```python
mu = 0 # 平均值
sigma = 1 # 标准差
X = np.arange(-4, 4, 0.01)
X
```

```python
# 概率密度函数 (PDF)

y = stats.norm.pdf(X,mu,sigma)
plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
fig,ax = plt.subplots(dpi=50)
plt.plot(X, y)
plt.xlabel('随机变量：x')
plt.ylabel('概率：y')
plt.title('正态分布：$\mu$=%.1f,$\sigma^2$=%.1f' % (mu,sigma))
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker

n = int(input("你要抛几次硬币？")) # 抛硬币的次数
p = 0.5 # 正面朝上的概率
X = np.arange(0, n+1,1) # 生成一个等差数组
print("X: ", X)

# 求对应分布的概率: 概率质量函数 (PMF)
# pmf 函数会返回一个列表，列表中每个元素表示随机变量对应的概率
pList = stats.binom.pmf(X, n, p)  # (X:正面朝上的次数，n:抛硬币的次数，p:正面朝上的概率)
print("pList: ", pList)

plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
fig,ax = plt.subplots(dpi=100)
plt.plot(X, pList, marker='o',linestyle='None')
plt.vlines(X, 0, pList,linestyles='dashed')
plt.xlabel('随机变量：抛硬币正面朝上次数')
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.ylabel('概率')
ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=1))
plt.title('二项分布：n=%i,p=%.2f' % (n,p))
plt.show()
```
